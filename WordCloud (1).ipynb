# Wordcloud

### Word Cloud to analyze word frequency in the first 4 chapter of Oliver Twist.

```python
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
text=open('OliverTwist.txt','r')
text=text.read()
```

```python
text
```

```python
stopwords=list(STOPWORDS)
stopwords.extend(('\n','said','the')) 
stopwords=set(stopwords)
```

```python
stopwords
```

```python
wc=WordCloud(width=1600, height=800,stopwords=stopwords)
```

```python
wc.generate(text)
```

```python
plt.figure( figsize=(20,10), facecolor='k')
plt.imshow(wc)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

## Masking over Images

Creating the wordcloud over an image of Oliver Twist

```python
image=Image.open('Oliver.JPG')
```

```python
image
```

```python
image=np.array(image)
```

```python
wc=WordCloud(width=1600, height=800,stopwords=stopwords,mask=image,background_color='white')
```

```python
wc.generate(text)
image_color=ImageColorGenerator(image)
```

```python
plt.figure( figsize=(20,10))
plt.imshow(wc.recolor(color_func=image_color),interpolation='bilinear')
plt.axis('off')
#plt.tight_layout(pad=0)
plt.show()
```
